Exploratory Data Analysis

```python
df.columns
```

```python
# EDA apply on Gender column
ax=sns.countplot(x='Gender',data=df)
```

```python
# lavels apply karne ke liye
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# groupby function use karte hai data ko group karne ke liye
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
# it means ki Gender ke basis pe Amount ko sum kar do and sort kar do Amount ke basis pe descending order me
```

```python
sales_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gender)
sns.set(rc={'figure.figsize':(7,5)})
```

from above graph  we can see the most of the buyers are females and even the purching power of females are greater then men

```python
# again EDA apply on Age group column
df.columns
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
#hue use karte hai ki alag alag color me dikhaye
for bars in ax.containers:
    ax.bar_label(bars)  
```

```python
df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

from above graph wee can see the most of the buyers are of age group betweem 26-35 years old female

```python
# EDA apply on state column
df.columns
```

```python
ax=sns.countplot(x='State',data=df)
sns.set(rc={'figure.figsize':(20,5)})
```

From above graph we can see that unexpectedly most opf the orders are from Uttar Pradesh, Maharashtra and Karnataka respectfully  but the total sales/amount is same
